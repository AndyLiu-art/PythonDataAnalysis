### 按照行名分组，针对多水平索引的情况

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
# 导入matplotlib包
import matplotlib as mpl
```

```python
df = pd.DataFrame(
    {
        "A": ["foo", "bar", "foo", "bar", "foo", "bar", "foo", "foo"],
        "B": ["one", "one", "two", "three", "two", "two", "one", "three"],
        "C": np.random.randn(8),
        "D": np.random.randn(8),
    }
)
df
```

```python
# 重新设置索引，将索引设置为A，B列的值
df2 = df.set_index(["A", "B"])
df2
```

```python
# 按照index的A进行分组
grouped = df2.groupby(level=df2.index.names.difference(["B"]))
# 对分组后的数据求和
grouped.sum()
```
