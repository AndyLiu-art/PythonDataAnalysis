### 固定dataframe的行名（列名不需要固定）

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
# 导入matplotlib包
import matplotlib as mpl
```

```python
# 设置随机数种子
np.random.seed(0)
# 生成数据
df = pd.DataFrame(np.random.randn(16, 100), columns=["A{}".format(i) for i in range(1, 101)])
df.style
```

```python
# 固定行名
df.style.set_sticky(axis="index")
```
