### 绘制数据的分组箱线图

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
```

```python
# 设置随机数种子
np.random.seed(123456)
# 生成数据
df = pd.DataFrame(np.random.rand(10, 2), columns=["Col1", "Col2"])
# 添加一列字符串，可以作为分类变量
df["X"] = pd.Series(["A", "A", "A", "A", "A", "B", "B", "B", "B", "B"])
df.plot(kind="box",vert=False, by="X")
```
