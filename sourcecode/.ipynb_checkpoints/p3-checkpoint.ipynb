### 以openml中的数据集diabetes为例，新建一个id列，划分为两个子数据集，两个数据集的样本量是一样的，只是变量不同。

### 将这两个的子数据集打乱顺序，按照id合并。

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
```

```python
# 获取数据
data = fetch_openml(
    data_id=37,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 新建一列id
data["id"] = range(1,data.shape[0]+1)
```

```python
# 按照preg排序，打乱顺序
data1 = data.loc[:, ["preg", "plas", "pres", "class", "id"]].sort_values(by=["preg"])
# 按照age排序，打乱顺序
data2 = data.loc[:, ["skin", "insu", "mass", "pedi", "age", "id"]].sort_values(by=["age"])
```

```python
data1.head()
```

```python
data2.head()
```

```python
# 按照id合并
newdata = data1.merge(data2, on="id")
# 按照id排序
newdata.sort_values(by=["id"], inplace=True)
newdata.head()
```
