### 以openml中的数据集CSM为例，计算评分Ratings前10的电影到底有多少人喜欢？

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
# 导入数据分析库
import pandas as pd
```

```python
# 获取数据
data = fetch_openml(
    data_id=42371,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 按照Rating降序排列，取出前10对应的Likes数
data.sort_values(by=["Ratings"], ascending=False).iloc[:10, ][["Ratings", "Likes"]]
```
