### 对含有缺失值的列或者行删除。

```python
import pandas as pd
import numpy as np
```

```python
# 生成dataframe
df = pd.DataFrame(
    np.random.randn(5, 3),
    index=["a", "c", "e", "f", "h"],
    columns=["one", "two", "three"],
)
df["four"] = "bar"
df["five"] = df["one"] > 0
df = df.reindex(["a", "b", "c", "d", "e", "f", "g", "h"])
df
```

```python
# 删除含有缺失值的行，存在缺失值就删除
df.dropna(axis=0, how="any")
```

```python
# 删除含有缺失值的列，全部都是缺失值才删除
df.dropna(axis=1, how="all")
```
