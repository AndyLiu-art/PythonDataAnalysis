### 对介于某两个数值之间的数据进行高亮显示（内置函数）

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
# 导入matplotlib包
import matplotlib as mpl
```

```python
# 设置随机数种子
np.random.seed(0)
# 生成数据
df = pd.DataFrame(np.random.randn(10,4), columns=["A","B","C","D"])
df.style
```

```python
left = pd.Series([1.0, 0.0, 1.0], index=["A", "B", "D"])
df.style.highlight_between(
    left=left, right=1.5, 
    axis=1, # 对列进行操作
    props="color:white; background-color:purple;"
)
```

```python
left = pd.Series([0, 0, 0, 0, 0], index=[0,2,3,5,8])
df.style.highlight_between(
    left=left, right=2.5, 
    axis=0, # 对行进行操作
    props="color:white; background-color:purple;"
)
```
