### 绘制多列数据的水平累积直方图，组数为20

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
```

```python
# 设置随机数种子
np.random.seed(123456)
# 生成数据
df = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 1,
    },
    columns=["a", "b", "c"]
)
df.plot(kind="hist", alpha=0.7, bins=20, orientation="horizontal", cumulative=True)
```
