### 以openml中的数据集1StudentPerformance为例，统计不同race和不同性别，不同的parental level of education下，写作成绩的方差。

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
# 导入数据分析库
import pandas as pd
```

```python
# 获取数据
data = fetch_openml(
    data_id=43255,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 数据透视表
pd.pivot_table(
    data,
    values="writing score",
    index=["gender", "parental level of education"],
    columns="race/ethnicity",
    aggfunc="var"
)
```
