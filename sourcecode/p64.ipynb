### 绘制多列数据的均值柱状图

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
```

```python
# 设置随机数种子
np.random.seed(123456)
# 生成数据
df = pd.DataFrame(
    np.random.randn(1000, 4), 
    index=pd.date_range("1/1/2000", periods=1000),
    columns=list("ABCD")
)
# 累加
df = df.mean()
df.plot(kind="bar")
```
