### 以openml中的数据集1StudentPerformance为例，找到math score小于reading score小于writing score的人。

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
# 导入数据分析库
import pandas as pd
```

```python
# 获取数据
data = fetch_openml(
    data_id=43255,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 修改列名，为了使用query语法
data.columns = data.columns[:5].to_list() + ["math_score", "reading_score", "writing_score"]
# 该查询方法直接是列名表达式
newdata = data.query("math_score < reading_score < writing_score")
newdata.head()
```
