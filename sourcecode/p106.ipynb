### 在dataframe上添加柱状图（内置函数）

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
# 导入matplotlib包
import matplotlib as mpl
```

```python
# 设置随机数种子
np.random.seed(0)
# 生成数据
df = pd.DataFrame(np.random.randn(10,4), columns=["A","B","C","D"])
df.style
```

```python
df.style.bar(subset=["A", "D"], color='#d65f5f')
```
