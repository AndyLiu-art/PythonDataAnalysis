### 绘制多列数据的水平柱状图，堆叠柱状图

```python
# 导入基础计算库
import numpy as np
# 导入数据分析库
import pandas as pd
```

```python
# 设置随机数种子
np.random.seed(123456)
# 生成数据
df = pd.DataFrame(
    np.random.randn(10, 4), 
    index=pd.date_range("1/1/2000", periods=10),
    columns=list("ABCD")
)
df.plot(kind="barh", stacked=True)
```
