### 以openml中的数据集1StudentPerformance为例，将宽数据转为长数据。

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
# 导入数据分析库
import pandas as pd
```

```python
# 获取数据
data = fetch_openml(
    data_id=43255,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 宽数据转为长数据
data.iloc[:5, [0,1,5,6]].melt(id_vars=["gender", "race/ethnicity"])
```
