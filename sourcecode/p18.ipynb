### 以openml中的数据集CSM为例，将数据集的部分列名进行修改。

```python
# 导入数据集获取工具
from sklearn.datasets import fetch_openml
# 导入数据分析库
import pandas as pd
```

```python
# 获取数据
data = fetch_openml(
    data_id=42371,
    as_frame=True,
    parser="pandas"
)["frame"]
```

```python
# 查看数据集的前五行
data.head()
```

```python
# 列重新命名
data.columns = [
    "年份",
    "评分",
    "类型",
    "gross",
    "预算",
    "screens",
    "续集",
    "感情",
    "观看次数",
    "喜欢人数",
    "不喜欢人数",
    "评论数",
    "粉丝数"
]
data.head()
```
